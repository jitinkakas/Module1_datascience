```python
"""
╔══════════════════════════════════════════════════════════════╗
║           ☕  COFFEE MACHINE - OOP PROJECT  ☕               ║
║                                                              ║
║  Concepts Used:                                              ║
║    • Abstraction   → Coffee (Abstract Base Class)            ║
║    • Inheritance   → All coffee types extend Coffee          ║
║    • Encapsulation → Recipe stored as private attributes     ║
║    • Polymorphism  → Each class overrides brew() & art()     ║
╚══════════════════════════════════════════════════════════════╝
"""

from abc import ABC, abstractmethod
import time


# ─────────────────────────────────────────────
#  ANSI COLOR CODES for terminal styling
# ─────────────────────────────────────────────
class Color:
    RESET   = "\033[0m"
    BOLD    = "\033[1m"
    RED     = "\033[91m"
    GREEN   = "\033[92m"
    YELLOW  = "\033[93m"
    BLUE    = "\033[94m"
    MAGENTA = "\033[95m"
    CYAN    = "\033[96m"
    WHITE   = "\033[97m"
    BROWN   = "\033[38;5;130m"
    ORANGE  = "\033[38;5;214m"
    LIME    = "\033[38;5;154m"
    CREAM   = "\033[38;5;229m"


# ─────────────────────────────────────────────
#  ABSTRACT BASE CLASS  →  Abstraction
# ─────────────────────────────────────────────
class Coffee(ABC):
    """
    Abstract class representing a generic Coffee.
    All subclasses MUST implement: brew() and art()
    """

    def __init__(self, name: str, temperature: str, serving_size: str):
        # Protected attributes (accessible in subclasses)
        self._name         = name
        self._temperature  = temperature
        self._serving_size = serving_size

    # Abstract methods — subclasses must override these
    @abstractmethod
    def brew(self) -> str:
        """Returns step-by-step brewing instructions."""
        pass

    @abstractmethod
    def art(self) -> list[str]:
        """Returns the hash (#) ASCII art as a list of strings."""
        pass

    @abstractmethod
    def color(self) -> str:
        """Returns the ANSI color code for this coffee's art."""
        pass

    # Concrete method shared by all (Inheritance benefit)
    def display(self):
        """Prints the full coffee card: art + recipe."""
        c = self.color()
        r = Color.RESET
        b = Color.BOLD

        print(f"\n{c}{'═' * 60}{r}")
        print(f"{b}{c}  ☕  {self._name.upper()}{r}")
        print(f"{c}{'═' * 60}{r}")

        # Print ASCII art
        print()
        for line in self.art():
            print(f"  {c}{line}{r}")

        # Print recipe/details
        print(f"\n{b}{Color.WHITE}  📋 RECIPE & DETAILS{r}")
        print(f"  {'─' * 40}")
        print(f"  🌡  Temperature : {self._temperature}")
        print(f"  🥤  Serving Size: {self._serving_size}")
        print(f"\n{b}{Color.WHITE}  👨‍🍳 BREWING STEPS:{r}")
        for i, step in enumerate(self.brew(), 1):
            print(f"  {c}{i}.{r} {step}")
        print(f"{c}{'═' * 60}{r}\n")

    # Getter methods — Encapsulation principle
    def get_name(self):         return self._name
    def get_temperature(self):  return self._temperature
    def get_serving_size(self): return self._serving_size


# ─────────────────────────────────────────────
#  SUBCLASS 1 → LATTE
# ─────────────────────────────────────────────
class Latte(Coffee):
    """
    Latte — milky espresso-based coffee.
    Encapsulates its own private recipe attributes.
    """

    def __init__(self):
        super().__init__("Latte", "65°C – 70°C", "240 ml")
        # Private recipe attributes (Encapsulation)
        self.__espresso_shots = 2
        self.__steamed_milk   = "180 ml"
        self.__milk_foam      = "1 cm thick layer"
        self.__syrup          = "Optional (vanilla/caramel)"

    def color(self): return Color.BROWN

    def art(self) -> list[str]:
        return [
            "        .######.",
            "      ##########",
            "     ###  __  ###      ╔══════════════╗",
            "    ###  /  \\  ###     ║  L A T T E   ║",
            "    ##  | () |  ##     ║  Creamy Bliss ║",
            "    ##  |    |  ##     ╚══════════════╝",
            "    ###  \\__/  ###",
            "     ###      ###",
            "      ##########",
            "        ######",
            "  ~~~~~~~~~~~~~~~~~~~~~~",
            "  |  espresso + milk  |",
            "  ~~~~~~~~~~~~~~~~~~~~~~",
        ]

    def brew(self) -> list[str]:
        return [
            f"Pull {self.__espresso_shots} shots of fresh espresso into your cup.",
            f"Steam {self.__steamed_milk} of whole milk to a velvety texture.",
            "Pour steamed milk slowly over espresso (creates layers).",
            f"Spoon {self.__milk_foam} of microfoam on top.",
            f"Add syrup if desired: {self.__syrup}.",
            "Serve immediately and enjoy! ☕",
        ]


# ─────────────────────────────────────────────
#  SUBCLASS 2 → MATCHA LATTE
# ─────────────────────────────────────────────
class MatchaLatte(Coffee):
    """
    Matcha Latte — Japanese green tea based drink.
    """

    def __init__(self):
        super().__init__("Matcha Latte", "60°C – 70°C", "300 ml")
        self.__matcha_powder = "2 teaspoons (ceremonial grade)"
        self.__hot_water     = "30 ml (not boiling)"
        self.__milk          = "250 ml (oat/almond/whole)"
        self.__sweetener     = "Honey or sugar to taste"

    def color(self): return Color.LIME

    def art(self) -> list[str]:
        return [
            "       .########.",
            "     ############",
            "    ##  ~~~~~~  ##      ╔══════════════════╗",
            "   ##  ~ MATCHA ~##     ║  M A T C H A     ║",
            "   ##  ~~  🍵  ~~~##    ║  Earthy & Smooth  ║",
            "   ##  ~~~~~~~~~~##     ╚══════════════════╝",
            "    ##  ~~~~~~  ##",
            "     ############",
            "       ########",
            "  ~~~~~~~~~~~~~~~~~~~~~~~~",
            "  |  matcha + steamed milk |",
            "  ~~~~~~~~~~~~~~~~~~~~~~~~",
        ]

    def brew(self) -> list[str]:
        return [
            f"Sift {self.__matcha_powder} into a bowl or cup.",
            f"Add {self.__hot_water} (80°C — NOT boiling, prevents bitterness).",
            "Whisk in a 'W' or 'M' motion until smooth and frothy.",
            f"Steam {self.__milk} until warm and silky.",
            "Pour steamed milk into the matcha mixture gently.",
            f"Sweeten with {self.__sweetener} and stir. 🍵",
        ]


# ─────────────────────────────────────────────
#  SUBCLASS 3 → AMERICANO
# ─────────────────────────────────────────────
class Americano(Coffee):
    """
    Americano — espresso diluted with hot water.
    """

    def __init__(self):
        super().__init__("Americano", "80°C – 85°C", "240 ml")
        self.__espresso_shots = 2
        self.__hot_water      = "150 ml"
        self.__ratio          = "1:3 (espresso:water)"

    def color(self): return Color.ORANGE

    def art(self) -> list[str]:
        return [
            "        .######.",
            "      ##########",
            "     ##  BOLD  ##      ╔══════════════════╗",
            "    ##  ~~~~~~  ##     ║  A M E R I C A N O║",
            "    ##   ~~~~   ##     ║  Strong & Clean   ║",
            "    ##  ~~~~~~  ##     ╚══════════════════╝",
            "    ##          ##",
            "     ##########",
            "       ######",
            "  ~~~~~~~~~~~~~~~~~~~~~~~~~",
            "  |  espresso + hot water |",
            "  ~~~~~~~~~~~~~~~~~~~~~~~~~",
        ]

    def brew(self) -> list[str]:
        return [
            f"Heat water to 92°C–96°C (just off boil).",
            f"Pull {self.__espresso_shots} shots of espresso into your cup.",
            f"Pour {self.__hot_water} of hot water into the cup.",
            f"Maintain ratio of {self.__ratio} for best balance.",
            "Do NOT add milk — keep it pure and strong! 💪",
            "Serve black or with a slice of lemon. 🍋",
        ]


# ─────────────────────────────────────────────
#  SUBCLASS 4 → CAPPUCCINO
# ─────────────────────────────────────────────
class Cappuccino(Coffee):
    """
    Cappuccino — equal parts espresso, steamed milk, and foam.
    """

    def __init__(self):
        super().__init__("Cappuccino", "65°C – 70°C", "180 ml")
        self.__espresso      = "60 ml (double shot)"
        self.__steamed_milk  = "60 ml"
        self.__foam          = "60 ml (thick & airy)"
        self.__cocoa_dust    = "Optional dusting on top"

    def color(self): return Color.MAGENTA

    def art(self) -> list[str]:
        return [
            "     .##########.",
            "   ##############",
            "  ##  ~~FOAM~~~  ##     ╔══════════════════╗",
            "  ##  ~~~~~~~~~  ##     ║  C A P P U C C I N O║",
            "  ##  ~~milk~~~  ##     ║  Frothy & Rich   ║",
            "  ##  ~~~~~~~~~  ##     ╚══════════════════╝",
            "  ##  espresso   ##",
            "   ##############",
            "     ##########",
            "  ~~~~~~~~~~~~~~~~~~~~~~~~~~",
            "  |  1/3 + 1/3 + 1/3 rule |",
            "  ~~~~~~~~~~~~~~~~~~~~~~~~~~",
        ]

    def brew(self) -> list[str]:
        return [
            f"Pull a double espresso shot: {self.__espresso}.",
            f"Steam {self.__steamed_milk} milk to silky microfoam (65°C).",
            f"Froth {self.__foam} separately (thicker than latte foam).",
            "Pour steamed milk into espresso first.",
            "Spoon thick foam generously on top.",
            f"Dust with cocoa or cinnamon: {self.__cocoa_dust}. ✨",
        ]


# ─────────────────────────────────────────────
#  SUBCLASS 5 → ESPRESSO
# ─────────────────────────────────────────────
class Espresso(Coffee):
    """
    Espresso — the pure, concentrated soul of coffee.
    """

    def __init__(self):
        super().__init__("Espresso", "88°C – 94°C", "30 ml")
        self.__coffee_grounds = "18–20g (fine grind)"
        self.__pressure       = "9 bars of pressure"
        self.__brew_time      = "25–30 seconds"
        self.__crema          = "Golden-orange crema on top"

    def color(self): return Color.RED

    def art(self) -> list[str]:
        return [
            "         .####.",
            "        ######",
            "       ##    ##        ╔════════════════╗",
            "       ##  ☕ ##       ║  E S P R E S S O║",
            "       ##    ##        ║  Pure & Intense  ║",
            "        ######         ╚════════════════╝",
            "         ####",
            "    ┌──────────┐",
            "    │ ~ crema ~│",
            "    │~~~~~~~~~~│",
            "    └──────────┘",
            "  ~~~~~~~~~~~~~~~~~~~~~",
            "  |  the soul of coffee |",
            "  ~~~~~~~~~~~~~~~~~~~~~",
        ]

    def brew(self) -> list[str]:
        return [
            f"Grind beans to fine powder: {self.__coffee_grounds}.",
            "Tamp the grounds with 30 lbs of even pressure.",
            f"Lock portafilter; brew at {self.__pressure}.",
            f"Extract for exactly {self.__brew_time}.",
            f"Look for {self.__crema} — sign of perfect extraction.",
            "Drink immediately — espresso waits for no one! ⚡",
        ]


# ─────────────────────────────────────────────
#  SUBCLASS 6 → COLD BREW
# ─────────────────────────────────────────────
class ColdBrew(Coffee):
    """
    Cold Brew — slow-steeped cold coffee concentrate.
    """

    def __init__(self):
        super().__init__("Cold Brew", "Cold / Iced", "350 ml")
        self.__coarse_grounds = "100g (coarse grind)"
        self.__cold_water     = "800 ml"
        self.__steep_time     = "12–24 hours in refrigerator"
        self.__ratio          = "1:8 (coffee:water)"

    def color(self): return Color.CYAN

    def art(self) -> list[str]:
        return [
            "       .##########.",
            "     ##############",
            "    ##   C O L D   ##   ╔══════════════════╗",
            "    ##   B R E W   ##   ║  C O L D  B R E W║",
            "    ## ❄ ~~ice~~ ❄ ##   ║  Smooth & Chill  ║",
            "    ## ~~~~~~~~~~~ ##   ╚══════════════════╝",
            "    ##  ░░░░░░░░░  ##",
            "     ##############",
            "       ##########",
            "    ~~~~~~~~~~~~~~~~~~~",
            "    |  slow & cold  |",
            "    ~~~~~~~~~~~~~~~~~~~",
        ]

    def brew(self) -> list[str]:
        return [
            f"Coarsely grind beans: {self.__coarse_grounds}.",
            f"Combine with {self.__cold_water} in a jar or pitcher.",
            f"Maintain ratio of {self.__ratio}.",
            "Stir gently, cover, and place in refrigerator.",
            f"Steep for {self.__steep_time} (longer = stronger).",
            "Strain through filter; serve over ice. ❄️",
        ]


# ─────────────────────────────────────────────
#  COFFEE MACHINE — Controller Class
# ─────────────────────────────────────────────
class CoffeeMachine:
    """
    Manages the coffee menu and user interaction.
    Uses polymorphism — calls display() on any Coffee subclass.
    """

    def __init__(self):
        # Menu: dict of option → Coffee object
        self.__menu: dict[str, Coffee] = {
            "1": Latte(),
            "2": MatchaLatte(),
            "3": Americano(),
            "4": Cappuccino(),
            "5": Espresso(),
            "6": ColdBrew(),
        }

    def __print_banner(self):
        print(f"\n{Color.BOLD}{Color.YELLOW}")
        print("  ╔══════════════════════════════════════════════════╗")
        print("  ║                                                  ║")
        print("  ║        ☕  WELCOME TO PYTHON CAFÉ  ☕            ║")
        print("  ║         Brewed with OOP Excellence               ║")
        print("  ║                                                  ║")
        print("  ╚══════════════════════════════════════════════════╝")
        print(f"{Color.RESET}")

    def __print_menu(self):
        print(f"{Color.BOLD}{Color.WHITE}  ┌─────────────────────────────┐")
        print("  │        📋  MENU             │")
        print("  ├─────────────────────────────┤")
        for key, coffee in self.__menu.items():
            name = coffee.get_name()
            size = coffee.get_serving_size()
            print(f"  │  [{key}]  {name:<15} {size:>8}  │")
        print("  │  [A]  Show ALL coffees       │")
        print("  │  [Q]  Quit                   │")
        print(f"  └─────────────────────────────┘{Color.RESET}")

    def __brewing_animation(self, name: str):
        c = Color.BROWN
        r = Color.RESET
        print(f"\n  {c}Brewing your {name}", end="", flush=True)
        for _ in range(5):
            time.sleep(0.3)
            print(f" .", end="", flush=True)
        print(f"  Done! ☕{r}\n")

    def run(self):
        self.__print_banner()

        while True:
            self.__print_menu()
            choice = input(
                f"\n  {Color.CYAN}Enter your choice: {Color.RESET}"
            ).strip().upper()

            if choice == "Q":
                print(f"\n{Color.YELLOW}  Thanks for visiting Python Café! ☕ Goodbye!{Color.RESET}\n")
                break

            elif choice == "A":
                print(f"\n{Color.BOLD}{Color.GREEN}  ── Displaying ALL coffees ──{Color.RESET}")
                for coffee in self.__menu.values():
                    self.__brewing_animation(coffee.get_name())
                    coffee.display()

            elif choice in self.__menu:
                selected = self.__menu[choice]
                self.__brewing_animation(selected.get_name())
                selected.display()

            else:
                print(f"\n  {Color.RED}  ✗ Invalid choice. Please try again.{Color.RESET}\n")


# ─────────────────────────────────────────────
#  ENTRY POINT
# ─────────────────────────────────────────────
if __name__ == "__main__":
    machine = CoffeeMachine()
    machine.run()
```
