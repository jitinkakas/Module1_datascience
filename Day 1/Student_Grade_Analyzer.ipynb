```python
import numpy as np

np.random.seed(42)
marks = np.random.randint(0, 100, size = 50)

print(marks)
```

```python
# Mean, Median, Lowest, Highest and Std 
mean = np.mean(marks)
median = np.median(marks)
lowest = np.min(marks)
highest = np.max(marks)
std_dev = np.std(marks)

print(f"Avg Marks : {mean}")
print(f"Median : {median}")
print(f"Lowest Score : {lowest}")
print(f"Highest Score : {highest}")
print(f"Std Deviation : {std_dev: .2f}")
```

```python
fail = marks[marks < 40]
dist = marks[marks > 85]
        
print(f"Failed Students : {len(fail)} -> Scores : {fail}")
print(f"Merit Students : {len(dist)} -> Scores : {dist}")   
```

```python
# Normalization between 0 to 1
normalized = (marks - lowest) / (highest - lowest)
print(f"Normalized : {np.round(normalized, 2)}")
```

```python
# Reshaped array
reshaped = marks.reshape(5, 10)
row_avg = np.mean(reshaped, axis = 1)
col_avg = np.mean(reshaped, axis = 0)

print(f"Reshaped : {reshaped}")
print(f"Avg of every subject : {np.round(row_avg, 2)}")
print(f"Avg of every student : {np.round(col_avg, 2)}")
```

```python
# Weighted Score
theory = marks[:25]
practical = marks[25:]

weighted_score = (0.4 * theory) + (0.6 * practical)
print(f"Weigted Score : {np.round(weighted_score, 2)}")
print(f"Weigted Avg Score : {np.mean(weighted_score): .2f}")
```
